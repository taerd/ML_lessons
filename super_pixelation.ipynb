```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
%matplotlib inline

N = 10
```

```python
image = imread('panda3.jpg')
plt.imshow(image);
```

```python
def img_to_rgbxy(img,alpha=1):
    xy= [[i*alpha,j*alpha] for i in range(img.shape[0]) for j in range(img.shape[1])]
    rgb = img.reshape((-1,image.shape[-1]))
    rgbxy = [np.concatenate([color,coords]) for color,coords in zip(rgb,xy)]
    return rgbxy
```

```python
X = img_to_rgbxy(image,1)
```

```python
kmeans = KMeans(n_clusters = N).fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
new_img = centroids[labels, 0:3].astype('int').reshape(image.shape)
plt.imshow(new_img);
```
