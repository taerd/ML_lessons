```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fake(y_min,y_max,x_min,x_max,h,image):
    hy = int((y_max-y_min)/h)
    hx = int((x_max-x_min)/h)
    for i in range(0,h):
        for j in range(0,h):
            image[(y_min+hy*i):(y_min+hy*(i+1)),\
                 (x_min+hx*j):(x_min+hx*(j+1)),0] = image[(y_min+hy*i):(y_min+hy*(i+1)),(x_min+hx*j):(x_min+hx*(j+1)),0].mean()
            image[(y_min+hy*i):(y_min+hy*(i+1)),\
                 (x_min+hx*j):(x_min+hx*(j+1)),1] = image[(y_min+hy*i):(y_min+hy*(i+1)),(x_min+hx*j):(x_min+hx*(j+1)),1].mean()
            image[(y_min+hy*i):(y_min+hy*(i+1)),\
                 (x_min+hx*j):(x_min+hx*(j+1)),2] = image[(y_min+hy*i):(y_min+hy*(i+1)),(x_min+hx*j):(x_min+hx*(j+1)),2].mean()
    return image
```

```python
image = imread('sam.jpg')
plt.imshow(image);
```

```python
fake_image = fake(150,500,250,550,10,image.copy())
plt.imshow(fake_image);
```
